```python
import math
import numpy as np
from random import randrange
import scipy.stats as stats
from keras.models import Sequential
from keras.layers import Dense, Activation
import pandas as pd
```

```python
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
def formula(n,x1,x2,score1,score2):
    a1 = np.array([3,3])
    a2 = np.array([3,-3])
    val = np.dot(np.array([x1[n],x2[n]]),a1)
    term1 = sigmoid(val)
    term2 = np.dot(a2,np.array([x1[n],x2[n]]))**2
    
    term3 = np.dot(np.array([0.3,0.3]),np.array([score1[n],score2[n]]))

    return (term1 + term2 + term3)
    
```

```python
x_train1=[]
x_train2=[]
y_train = []
x_test1=[]
x_test2=[]
y_test = []
```

```python
def generate_data():
    for i in range(10000):
        x_train1.append(randrange(0,7001)/10000)
    for j in range(10000):
        x_train2.append(randrange(0,7001)/10000)
    for i in range(3000):
        x_test1.append(randrange(7000,10001)/10000)
    for j in range(3000):
        x_test2.append(randrange(7000,10001)/10000)
    score1 = stats.zscore(np.array(x_train1))
    score2 = stats.zscore(np.array(x_train2))
    score3 = stats.zscore(np.array(x_test1))
    score4 = stats.zscore(np.array(x_test2))
    for i in range (10000):
        y_train.append(formula(i,x_train1,x_train2,score1,score2).item())
    for i in range (3000):
        y_test.append(formula(i,x_test1,x_test2,score3,score4).item())
    
```

```python
generate_data()
```

```python
x_train = np.vstack((x_train1,x_train2))
x_train = pd.DataFrame(x_train)
y_train =  pd.DataFrame(y_train)
y_train = y_train.transpose()
y_test =  pd.DataFrame(y_test)
y_test = y_test.transpose()
x_test = np.vstack((x_test1,x_test2))
x_test = pd.DataFrame(x_test)
```

```python
x_train = x_train.transpose() 
y_train = y_train.transpose() 
x_test = x_test.transpose() 
y_test = y_test.transpose() 
```

```python
from keras import backend as K
def root_mean_squared_error(y_true, y_pred):
        return K.sqrt(K.mean(K.square(y_pred - y_true))) 
```

```python
model = Sequential()

model.add(Dense(10,input_dim=2, activation='sigmoid'))
model.add(Dense(1, activation='linear'))

# Compile the model and calculate its accuracy:
model.compile(loss=root_mean_squared_error) 
model.summary()
model.fit(x_train, y_train, epochs=100, batch_size=64,validation_data=(x_test, y_test))
# Print a summary of the Keras model:
```

```python
import keras
from matplotlib import pyplot as plt
history = model.history
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
min = val_loss[99]-loss[99]
```

# seraching for best number of hidden layers

```python
best_model = model
model1=[]
model1.append(model)
for i in range(1,10):
    model1.append(Sequential())
    model1[i].add(Dense(10,input_dim=2, activation='sigmoid'))
    for j in range(i):
        print(i)
        model1[i].add(Dense(10, activation='sigmoid')) 
    
    model1[i].add(Dense(1, activation='linear'))

    # Compile the model and calculate its accuracy:
    model1[i].compile(loss=root_mean_squared_error) 
    model1[i].summary()
    model1[i].fit(x_train, y_train, epochs=100, batch_size=64,validation_data=(x_test, y_test))
    history = model1[i].history
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    if (val_loss[99]-loss[99])<min:
        min = (val_loss[99]-loss[99])
        best_model = model1[i]
```

```python
history = best_model.history
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

# early stopping

```python
from keras.callbacks import EarlyStopping
model = Sequential()

model.add(Dense(10,input_dim=2, activation='sigmoid'))
model.add(Dense(10 ,activation='sigmoid'))
model.add(Dense(1, activation='linear'))

# Compile the model and calculate its accuracy:
model.compile(loss=root_mean_squared_error) 
model.summary()
earlystop_callback = EarlyStopping(monitor='val_loss',min_delta=0.05,patience=1)
model.fit(x_train, y_train, epochs=100, batch_size=64,validation_data=(x_test, y_test),callbacks=[earlystop_callback])
```

```python
history = model.history
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
